```python
csv_paths = []
categories = []
import os
for root, dirs, files in os.walk("data/Chinese-medical-dialogue-data-master/Data_数据"):
    # print(root, dirs, files)
    categories += dirs
    for f in files:
        if f.endswith(".csv"):
            csv_paths.append(os.path.join(root, f))
print(csv_paths)
categories
```

```python
frames = []
for ind, csv_path in enumerate(csv_paths):
    frame = pd.read_csv(csv_path,encoding="gb18030")
    frame["category"] = [categories[ind]] * len(frame)
    frames.append(frame)
all_data = pd.concat(frames)
all_data
```

```python
all_data.groupby("category").count()
```

```python
all_data.to_csv("data/Chinese-medical-dialogue-data-master/all_data.csv", encoding="utf-8")
```
